
- Lists in python are mutable
- Some functions
    - `len()` will provide the length of a list
    - `sum()` will sum the values in the list
    - `max()` will return the highest value in a list of numerics
    - `min()` will return the smallest value in a list of numerics

## Generators

- Generators are functions returning an iterable instead of a sequence
- To declare a generator, declare a function using a `yield` statement instead of a `return` statement
- A `return` statement terminates a function entirely, while a `yield` statement pauses the function, saves all its states and later continues from there on successive calls.
- At the end of the iteration, the generator is empty. 
- Generators are an efficient way to optimizing memory
- Generators can also be called for all values using `list()` instead of `next()`

````python
def squaree(list_of_x):
    for x in list_of_x:
        yield(x**2)

to_generate = squaree([4,5,6])
next(to_generate)
>>16

next(to_generate)
>>25

next(to_generate)
>>36

list(to_generate)
>>[16, 25, 36]

to_generate = (i**2 for i in [4,5,6])
next(to_generate)
>>16

list(to_generate)
>>[25, 36]

def generatorEvenOdd2(m: list):
    for i in m:
        yield("even" if i%2==0 else "odd")
list(generatorEvenOdd2([1,2,3,4,5,6]))
>>['odd', 'even', 'odd', 'even', 'odd', 'even']

entry = [1, 2, 3]
gen = (x**2 for x in entry)
gen
>><generator object <genexpr> at 0x000001C0C2667510>
list(gen)
>>[1, 4, 9, 16, 25, 36, 49, 64, 81]
list(gen)
>>[]

````

### re.sub
`re.sub` is a function in Python's `re` module (regular expressions) used for searching and replacing patterns in strings. Here's a concise overview of its arguments and purpose:

**Usage : **

```python
re.sub(pattern, repl, string, count=0, flags=0)

tweet = re.sub(r'\$\w*', '', tweet)
```

**Arguments:**

- **pattern**: The regular expression pattern to search for within the string.
- **repl**: The replacement string. Can be either a string or a function; if it's a function, it is called for every non-overlapping occurrence of the pattern.
- **string**: The string to search and replace within.
- **count**: (Optional) The maximum number of pattern occurrences to replace. 0 means replace all occurrences.
- **flags**: (Optional) Modifier flags to change how the search is performed, such as `re.IGNORECASE`.

**Purpose:**

`re.sub` is used for:

- **Searching and Replacing**: Finds all occurrences of the pattern in the string and replaces them with `repl`.
- **Data Cleaning**: Useful in data preprocessing to remove or replace unwanted characters or patterns.
- **Text Transformation**: Can be used to modify text in a file or input by applying specific patterns and replacements.



## zip()

 - The **zip()** function in Python creates an iterator that aggregates elements from two or more iterables (lists, tuples, sets, etc.). It's commonly used to loop over multiple sequences simultaneously, with corresponding elements from each sequence being paired together.

```plaintext
    zip(*iterables)
```

 - **`*iterables`**: One or more sequences of elements (e.g., lists, tuples).

**Example**

`````python
yslist = [1, 2, 3]
tweets = ["Tweet 1", "Tweet 2", "Tweet 3"]

for y, tweet in zip(yslist, tweets):
    print(f"Data: {y}, Tweet: {tweet}")

# Output
# Data: 1, Tweet: Tweet 1
# Data: 2, Tweet: Tweet 2
# Data: 3, Tweet: Tweet 3
`````

`````python
questions = ['name', 'quest', 'favorite color']
answers = ['lancelot', 'the holy grail', 'blue']
for q, a in zip(questions, answers):
    print('What is your {0}?  It is {1}.'.format(q, a))

# What is your name?  It is lancelot.
# What is your quest?  It is the holy grail.
# What is your favorite color?  It is blue.
`````

## Inverser l'ordre des éléments d'une liste 

L'expression `idx_sorted[::-1]` est utilisée pour inverser l'ordre des éléments d'une liste ou d'un tableau en Python. Expliquons cette expression en détail :

- `idx_sorted` est une variable qui contient une liste ou un tableau. Les éléments de cette liste ou de ce tableau sont supposés être déjà triés dans un certain ordre (croissant, par exemple).

- L'opération de slicing (découpage) `[::-1]` est appliquée à `idx_sorted`. Voici ce que signifient les différents éléments de cette opération de slicing :
  - Le premier `:` indique que nous commençons du premier élément jusqu'au dernier élément de la liste ou du tableau. En l'absence de nombre avant ou après ce symbole, il est sous-entendu que nous incluons tous les éléments de la séquence.
  - Le deuxième `:` est utilisé pour séparer le début et la fin de l'intervalle du pas de l'itération.
  - `-1` indique le pas de l'itération. Un pas de `-1` signifie que nous parcourons la liste ou le tableau en arrière, d'où l'effet d'inversion.

En résumé, `idx_sorted[::-1]` prend tous les éléments de `idx_sorted` depuis le début jusqu'à la fin (puisque ni le début ni la fin ne sont spécifiés) mais avec un pas de `-1`, ce qui fait que la liste ou le tableau est parcouru en sens inverse, résultant en une liste ou un tableau où l'ordre des éléments est inversé par rapport à l'ordre original.

Par exemple, si `idx_sorted = [1, 2, 3, 4, 5]`, alors `idx_sorted[::-1]` donnera `[5, 4, 3, 2, 1]`.
